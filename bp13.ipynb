```python
import math
c=50
h=30
value = []
items=[x for x in input().split(',')]
for d in items:
    value.append(str(int(round(math.sqrt(2*c*float(d)/h)))))

print(','.join(value))
```

```python
row=int(input("Enter the number of rows"))
coloumn=int(input("Enter the number of coloumns"))
matrix=[[0 for j in range(coloumn)]for i in range(row) ]
for i in range(row):
    for j in range(coloumn):
        matrix[i][j]=i*j
matrix
```

```python
a="hello world! 123"

letters=0
digits=0
for i in a:
    if i.isdigit():
        digits+=1
    elif i.isalpha():
        letters+=1
    else:
        pass
print("LETTERS",letters)
print("DIGITS",digits)
```

```python
import re
password=[i for i in input("Enter the passwords:").split(",")]
reg="^(?=.*[a-z])(?=.*[A-Z])(?=.*[0-9])(?=.*[$#@])[a-zA-Z0-9$#@]{6,12}$"
match_re=re.compile(reg)
for i in password:
    res=re.search(reg,i)
    if res:
        print("Valid Password",i)
```
